```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

class StudentDataset:
    """Generates synthetic student data."""

    def __init__(self, num_students, seed=None):
        """
         Initializes the StudentDataset.

        Args:
            num_students: The number of student records to generate.
            seed: An optional random seed for reproducibility.
        """
        self.num_students = num_students
        self.seed = seed
        self.rng = np.random.default_rng(seed)
        self.df = None # Initialize DataFrame attribute

    def generate_gender(self):
        """Generates genders based on a specified distribution."""
        gender_distribution = {'Male': 0.5, 'Female': 0.5}
        genders = self.rng.choice(list(gender_distribution.keys()),
                                  size=self.num_students,
                                  p=list(gender_distribution.values()))
        return genders

    def generate_major(self):
        """Generates majors based on a specified distribution."""
        major_distribution = {'Computer Science': 0.5, 'Physics': 0.2, 'Chemistry': 0.2, 'Mathematics': 0.1}
        majors = self.rng.choice(list(major_distribution.keys()),
                                 size=self.num_students,
                                 p=list(major_distribution.values()))
        return majors

    def generate_program(self, majors):
        """Generates programs conditioned on the generated majors."""
        program_distributions = {
            'Computer Science': {'BS in CS': 0.7, 'BA in CS': 0.3},
            'Physics': {'BS in Physics': 0.6, 'BA in Physics': 0.4},
            'Chemistry': {'BS in Chemistry': 0.65, 'BA in Chemistry': 0.35},
            'Mathematics': {'BS in Mathematics': 0.55, 'BA in Mathematics': 0.45}
        }
        programs = []
        for major in majors:
            distribution = program_distributions.get(major, {})
            if not distribution:
                programs.append(None) # Handle cases where major is not in the distribution
            else:
                program = self.rng.choice(list(distribution.keys()), p=list(distribution.values()))
                programs.append(program)
        return programs

    def generate_gpa(self, majors):
        """Generates GPAs normally distributed by major and clipped."""
        gpa_parameters = {
            'Computer Science': {'mean': 3.5, 'std': 0.3},
            'Physics': {'mean': 3.6, 'std': 0.25},
            'Chemistry': {'mean': 3.4, 'std': 0.35},
            'Mathematics': {'mean': 3.7, 'std': 0.2}
        }
        gpas = []
        for major in majors:
            params = gpa_parameters.get(major, {'mean': 3.0, 'std': 0.5}) # Default if major not found
            gpa = self.rng.normal(loc=params['mean'], scale=params['std'])
            clipped_gpa = np.clip(gpa, 0.0, 4.0)
            gpas.append(clipped_gpa)
        return gpas


    def assemble_dataframe(self):
        """Assembles the full DataFrame by generating each attribute."""
        genders = self.generate_gender()
        majors = self.generate_major()
        programs = self.generate_program(majors)
        gpas = self.generate_gpa(majors)

        data = {'Gender': genders, 'Major': majors, 'Program': programs, 'GPA': gpas}
        self.df = pd.DataFrame(data)


    def get_full_dataframe(self):
        """
        Returns the assembled DataFrame.

        If the DataFrame has not been assembled yet, it will call
        assemble_dataframe first.

        Returns:
            A pandas DataFrame containing the generated student data.
        """
        if self.df is None:
          self.assemble_dataframe()

        return self.df

    def plot_gender_distribution(self):
        """Visualizes the distribution of genders."""
        if self.df is None:
            self.assemble_dataframe()

        gender_counts = self.df['Gender'].value_counts()

        plt.figure(figsize=(6, 4))
        sns.barplot(x=gender_counts.index, y=gender_counts.values)
        plt.title('Distribution of Student Genders')
        plt.xlabel('Gender')
        plt.ylabel('Number of Students')
        plt.show()

    def plot_major_distribution(self):
        """Visualizes the distribution of majors."""
        if self.df is None:
            self.assemble_dataframe()

        major_counts = self.df['Major'].value_counts()

        plt.figure(figsize=(10, 6))
        sns.barplot(x=major_counts.index, y=major_counts.values)
        plt.title('Distribution of Student Majors')
        plt.xlabel('Major')
        plt.ylabel('Number of Students')
        plt.show()

    def plot_program_distribution(self):
        """Visualizes the distribution of programs."""
        if self.df is None:
            self.assemble_dataframe()

        program_counts = self.df['Program'].value_counts()

        plt.figure(figsize=(12, 6))
        sns.barplot(x=program_counts.index, y=program_counts.values)
        plt.title('Distribution of Student Programs')
        plt.xlabel('Program')
        plt.ylabel('Number of Students')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout() # Adjust layout to prevent labels overlapping
        plt.show()

    def plot_gpa_distribution(self):
        """Visualizes the overall GPA distribution."""
        if self.df is None:
            self.assemble_dataframe()

        plt.figure(figsize=(8, 6))
        sns.histplot(data=self.df, x='GPA', kde=True, bins=20)
        plt.title('Distribution of Student GPAs')
        plt.xlabel('GPA')
        plt.ylabel('Frequency')
        plt.show()

    def plot_program_by_major(self):
        """Visualizes the program distribution conditioned on major."""
        if self.df is None:
            self.assemble_dataframe()

        plt.figure(figsize=(14, 8))
        sns.countplot(data=self.df, x='Program', hue='Major')
        plt.title('Program Distribution by Major')
        plt.xlabel('Program')
        plt.ylabel('Number of Students')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()

    def plot_gpa_by_major(self):
        """Visualizes the GPA distribution conditioned on major."""
        if self.df is None:
            self.assemble_dataframe()

        plt.figure(figsize=(10, 6))
        sns.boxplot(data=self.df, x='Major', y='GPA')
        plt.title('GPA Distribution by Major')
        plt.xlabel('Major')
        plt.ylabel('GPA')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()

    def plot_gpa_by_program(self):
        """Visualizes the GPA distribution conditioned on program."""
        if self.df is None:
            self.assemble_dataframe()

        plt.figure(figsize=(12, 6))
        sns.boxplot(data=self.df, x='Program', y='GPA')
        plt.title('GPA Distribution by Program')
        plt.xlabel('Program')
        plt.ylabel('GPA')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()

    def plot_gpa_by_program_and_major(self):
        """Visualizes the GPA distribution conditioned on program and major."""
        if self.df is None:
            self.assemble_dataframe()

        plt.figure(figsize=(16, 8)) # Consider a suitable figure size
        sns.boxplot(data=self.df, x='Program', y='GPA', hue='Major')
        plt.title('GPA Distribution by Program and Major')
        plt.xlabel('Program')
        plt.ylabel('GPA')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()

    def plot_sampled_dataset(self):
        """Visualizes the gender, major, program, and GPA of 100 randomly sampled students."""
        if self.df is None:
            self.assemble_dataframe()

        sampled_df = self.df.sample(n=100, random_state=self.seed)

        sns.pairplot(sampled_df, hue='Major')
        plt.suptitle('Sampled Student Data Distribution (100 Students)', y=1.02)
        plt.show()

    def plot_entire_dataset_summary(self):
        """Generates a summary visualization of the entire dataset (e.g., pairplot)."""
        if self.df is None:
            self.assemble_dataframe()

        sns.pairplot(self.df, hue='Major')
        plt.suptitle('Entire Student Dataset Summary', y=1.02)
        plt.show()

    def gpa_mean_std(self):
        """Calculates the mean and standard deviation of the GPA."""
        self.get_full_dataframe()  # Ensure DataFrame is assembled
        gpa_mean = self.df['GPA'].mean()
        gpa_std = self.df['GPA'].std()
        return gpa_mean, gpa_std

    def count_students_per_program_major_pair(self):
        """Counts the number of students for each unique (program, major) pair."""
        self.get_full_dataframe() # Ensure DataFrame is assembled
        program_major_counts = self.df.groupby(['Program', 'Major']).size().reset_index(name='Count')
        return program_major_counts

    def visualize_students_per_program_major_pair(self):
        """Visualizes the counts of students per program-major pair using a heatmap."""
        program_major_counts = self.count_students_per_program_major_pair()
        pivot_table = program_major_counts.pivot(index='Program', columns='Major', values='Count').fillna(0).astype(int)

        plt.figure(figsize=(10, 6))
        sns.heatmap(pivot_table, annot=True, fmt='d', cmap='viridis')
        plt.title('Number of Students per Program and Major Pair')
        plt.xlabel('Major')
        plt.ylabel('Program')
        plt.show()

student_data = StudentDataset(num_students=10000, seed=42)
program_major_counts = student_data.count_students_per_program_major_pair()
student_data.visualize_students_per_program_major_pair()
```
